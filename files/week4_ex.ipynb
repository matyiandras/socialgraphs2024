Regex

- regex is a special sequence of characters that helps you match or find other strings or sets of strings, using a specialized syntax held in a pattern.
- four digits: '\W\d\d\d\d\W'
- words starting with super: 'super\w*'
- wikipedia links: '\[\[(.*?)(?:\|([^\|]*?))?\]\]'

```python
#read musician_list file

import re

#open file
file = open("musician_list.txt", "r")
text = file.read()
file.close()

musician_links = [x[0].replace(" ", "_") for x in re.findall(r"\[\[(.*?)(?:\|([^\|]*?))?\]\]", text)]
print(musician_links)
```

```python
import networkx as nx

MUSICIANS = nx.DiGraph()

for musician in musician_links:
    MUSICIANS.add_node(musician)

baseurl = "https://en.wikipedia.org/w/api.php?"
action = "action=query"

content = "prop=revisions&rvprop=content"
dataformat ="format=json"

# titles = "titles={}".format(musician_links[1])
# query = "{}{}&{}&{}&{}".format(baseurl, action, content, titles, dataformat)

# wikiresponse = urllib.request.urlopen(query)
# wikidata = wikiresponse.read()
# wikitext = wikidata.decode('utf-8')

# print(wikitext)

# data = json.loads(wikitext)
# content = data['query']['pages'][list(data['query']['pages'].keys())[0]]['revisions'][0]['*']

# print(content)

print(musician_links)

for musician in musician_links:

    print(musician)

    titles = "titles={}".format(musician)
    query = "{}{}&{}&{}&{}".format(baseurl, action, content, titles, dataformat)

    print(query)

    wikiresponse = urllib.request.urlopen(query)
    wikidata = wikiresponse.read()
    wikitext = wikidata.decode('utf-8')

    data = json.loads(wikitext)
    page_content = data['query']['pages'][list(data['query']['pages'].keys())[0]]['revisions'][0]['*']

    if page_content.find("REDIRECT") != -1:
        redirect = re.search(r"\[\[(.*?)(?:\|([^\|]*?))?\]\]", page_content).group(1).replace(" ", "_")
        print(redirect)
        nx.relabel_nodes(MUSICIANS, {musician: redirect}, copy=False)

print(MUSICIANS.nodes())



```
